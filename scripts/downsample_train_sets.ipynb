```python

from hist_smell.evaluation.near_misses import read_pred_file

mt_train_file = 'test-files-mt-all-lang/out-fold-smell-en/folds_2_train.tsv'

train_tokens = list(read_pred_file(mt_train_file))
len(train_tokens)
```

```python
from collections import defaultdict

sent_tokens = defaultdict(list)
for token in train_tokens:
    sent = (token.text_id, token.sent_idx)
    sent_tokens[sent].append(token)
```

```python
import random

import numpy as np

from hist_smell.evaluation.near_misses import write_pred_file

random_seed = 2629

num_sents = len(sent_tokens)
num_sents
fractions = np.arange(0.2, 1.0, 0.2)
for fraction in fractions:
    num_sents_frac = int(fraction * num_sents)
    random.seed(random_seed)
    frac_sents = random.sample(list(sent_tokens.keys()), num_sents_frac)
    #print(fraction, len(frac_sents), frac_sents[:10])
    frac_file = mt_train_file.replace("train.tsv", f"train-train_size_{fraction:.1f}.tsv")
    frac_tokens = [token for sent in frac_sents for token in sent_tokens[sent]]
    write_pred_file(frac_file, frac_tokens)
    
```
